```python
#choisir au hasard 5 joueurs qui seront utilisés comme centroides initiaux

num_clusters = 5

random_initial_points = np.random.choice(point_guards.index, size = num_clusters )

centroides = point_guards.loc[random_initial_points]
```

```python
#Dictionnaire de centroides [important pour voir les points sur la figure précéd]

def centroides_to_dict(centroides) : 
    dictionary = dict()
    counter=0
    
    for i,row in centroides.iterrows():
        coordinates = [row['ppg'], row['atr']]
        dictionary[counter] = coordinates
        counter +=1
    return dictionary

centroides_dict = centroides_to_dict(centroides)
centroides_dict
```

```python
#calcul de la ditance de chaque joueur au centroide

import math

def calculate_distance(centroid,player_values) : 
    root_distance = 0
    
    for x in range(0,len(centroid)) : 
        
        diff = centroid[x] - player_values[x]
        
        squarred_difference = diff ** 2
        
        root_distance += squarred_difference
        
        euclid_distance = math.sqrt(root_distance)
        
    return euclid_distance

```

## Etape 1 : 

```python
#Fonction qui asigne une ligne à un cluster
def assign_to_cluster(row) : 
    
    low_dist = -1
    closest_clust = -1 
    
    for cluster_id, centroid in centroides_dict.items() : 
        df_row = [row['ppg'], row['atr']]

        distance = calculate_distance(centroid,df_row)

        if low_dist == -1 : 
            low_dist = distance
            closest_clust = cluster_id
        elif low_dist > distance : 
            low_dist = distance
            closest_clust = cluster_id
            
    return closest_clust

```

```python
#Appliquer la fonction au jeu de données pour créer une colomne 
# cluster:

point_guards['cluster'] = point_guards.apply(lambda x: assign_to_cluster(x), axis=1)

point_guards
```
