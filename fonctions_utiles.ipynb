# Overfitting
### to test if there is overfitting try the following functions : 
#### U can start by using the computation of the 1 col, 2, 3... etc 
#### U should have the following relationship 



```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

def train_test(data,cols) : 
    
    lr = LinearRegression()
    
    target = data.target
    
    lr.fit(data[cols],target)
    
    predictions = lr.predicts(data[cols])
    
    
    mse= mean_squared_error(target, predictions)
    variance = np.var(predictions)
    
    return(mse,variance)
```

#### Then you should apply the train_and_cross_val and compare the results to the previous results (actually u can scatter the results)



```python
from sklearn.model_selection import KFold 

def train_and_cross_val(data,cols) : 
    mse_values =[]
    var_values =[]
    target = data.target
    
    kf = KFold(n_splits=10, shuffle=True, random_state=3)
    
    for train_index, test_index in kf.split(data[cols]) : 
        X_train, X_test = data[cols].iloc[train_index], data[cols].iloc[test_index]
        y_train, y_test = target.iloc[train_index], target.iloc[test_index]
        
        lr = LinearRegression()
        
        lr.fit(X_train, y_train)
        predictions = lr.predict(X_test)
        
        mse= mean_squared_error(y_test, predictions)
        variance = np.var(predictions)
        
        mse_values.append(mse)
        var_values.append(variance)
        
    avg_mse = np.mean(mse_values)
    avg_var = np.mean(var_values)
    
    return (avg_mse, avg_var)


#to trace the results
#import matplotlib.pyplot as plt
#plit.scatter([2,3,..],[two_rmse,three_rmse..],c='blue')
#plt.show()




    
```
